```python
import pandas as pd
import folium

Map_df = pd.read_csv('nuclear_explosions.csv')
HEAT_MAP_df = Map_df[["Location.Cordinates.Latitude", "Location.Cordinates.Longitude", "WEAPON DEPLOYMENT LOCATION"]]
HEAT_MAP_df.head(10)
```

```python
plotly_plot = folium.Map(location=[0,0], tiles="cartodbpositron", zoom_start=2)
for i in range(0, len(HEAT_MAP_df)):
    folium.Marker(location= [HEAT_MAP_df.iloc[i]['Location.Cordinates.Latitude'], HEAT_MAP_df.iloc[i]['Location.Cordinates.Longitude']],
                          popup= HEAT_MAP_df.iloc[i]['WEAPON DEPLOYMENT LOCATION'],).add_to(plotly_plot)
plotly_plot
```
